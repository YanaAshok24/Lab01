#### Data Cleaning

```python
df.isna().sum()
```

It's clear that there are 2 unwanted columns and several rows which have Null values. Removing them

```python
df.dropna(how='all', axis=1, inplace=True)
df
```

```python
df.dropna(how = 'all', axis = 0, inplace = True)
df
```

Now the dataset has only 9357 rows and 15 columns

Replacing all comma to dot and coverting the columns to float values

```python
df["CO(GT)"] = df["CO(GT)"].str.replace(",", ".").astype(float)
df["C6H6(GT)"] = df["C6H6(GT)"].str.replace(",", ".").astype(float)
df["T"] = df["T"].str.replace(",", ".").astype(float)
df["RH"] = df["RH"].str.replace(",", ".").astype(float)
df["AH"] = df["AH"].str.replace(",", ".").astype(float)
df["Time"] = df["Time"].str.replace(".", ":")
df
```

Changing to datetime format

```python
df.Date = pd.to_datetime(df.Date)
df.Time = pd.to_datetime(df.Time)
df['Time'] = pd.Series([val.time() for val in df['Time']])
df
```
